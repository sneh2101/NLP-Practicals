```python

#function for counting number of times one tag occurs, no. of times a word given a tag occurs
def calc_emission(word, tag, train_bag = tagged_corpus):
    tag_list = [pair for pair in train_bag if pair[1]==tag]
    count_tag = len(tag_list)#total number of times the passed tag occurred in the tagged_corpus
    w_given_tag_list = [pair[0] for pair in tag_list if pair[0]==word] #makes a list of all the occurences of a specific word given a particular tag
#now calculate the total number of times the passed word occurred as the passed tag.
    count_w_given_tag = len(w_given_tag_list)

    
    return (count_w_given_tag, count_tag)
```

```python
small_text_words = text_words[:20]
print(small_text_words)
```

```python
# creating w x t emission matrix of words and tags, w= no of words, t= no of tags


emission_matrix = np.zeros((20,27), dtype='float32') #taking a small sample of the first 20 words from the corpus for convenience
for i, word in enumerate(small_text_words):
    for j, tag in enumerate(list(tags)): 
        emission_matrix[i, j] = calc_emission(word,tag )[0]/calc_emission(word, tag)[1] #emission_probability(P(w/t)) = no. w given tag / no. of tag 

print(emission_matrix)
```

```python
emission_df = pd.DataFrame(emission_matrix, columns = list(tags), index=list(small_text_words))
display(emission_df) #creating dataframe for convenience
```

```python
#calculating transition probability
def t2_given_t1(t2, t1, train_bag = tagged_corpus):
    tags = [pair[1] for pair in train_bag] #list of tags in the corpus
    count_t1 = len([t for t in tags if t==t1]) #no of times a tag t1 occurs
    count_t2_t1 = 0
    for index in range(len(tags)-1):
        if tags[index]==t1 and tags[index+1] == t2: 
            count_t2_t1 += 1 #no of times a tag t1 occurs right before another tag t2
    return (count_t2_t1, count_t1)
```

```python
# creating t x t transition matrix of tags, t= no of tags


tags_matrix = np.zeros((len(tags), len(tags)), dtype='float32')
for i, t1 in enumerate(list(tags)):
    for j, t2 in enumerate(list(tags)): 
        tags_matrix[i, j] = t2_given_t1(t2, t1)[0]/t2_given_t1(t2, t1)[1] # transmission_probability (P(t1t2/t1))= no. of times a tag t1 offucrs before tag t2/no. of times tag t1 occurs 

print(tags_matrix)
```

```python
# convert the matrix to a df for better readability

tags_df = pd.DataFrame(tags_matrix, columns = list(tags), index=list(tags))
display(tags_df)
```

```python
#viterbi algorithm to predict the mist suitable  pos tags
def Viterbi(words, train_bag = tagged_corpus):
    state = []
    T = list(set([pair[1] for pair in train_bag]))
    
    for key, word in enumerate(words):
        #initialise list of probability column for a given observation
        p = []  #list fior storing state probability
        for tag in T:
            if key == 0:
                transition_p = tags_df.loc['.', tag]
            else:
                transition_p = tags_df.loc[state[-1], tag]
                
            # compute emission and state probabilities
            emission_p = calc_emission(words[key], tag)[0]/calc_emission(words[key], tag)[1] #emission_probability(P(w/t)) = no. w given tag / no. of tag 
            state_probability = emission_p * transition_p    
            p.append(state_probability)
            
        pmax = max(p)
        # getting state for which probability is maximum
        state_max = T[p.index(pmax)] 
        state.append(state_max) #list of final most probable states 
    return list(zip(words, state))
```

```python
ans = Viterbi(text_words)
print(ans)
```

```python
from tabulate import tabulate

formatted_ans = []
for i in range(len(tagged_corpus)):
  formatted_ans.append([tagged_corpus[i][0], ans[i][1], tagged_corpus[i][1]])

print(tabulate(formatted_ans, headers= ["Word", "Predicted Tag", "Actual Tag"]))

#creating table of the corpus, predicted tag as predicted by the HMM model using the Viterbi Algorithm, Actual Tag Provided by Spacy
```
